```python
# Create a new DataFrame with 10 more people
new_data = {
    'Name': ['John', 'Emily', 'Michael', 'Sophia', 'William', 'Olivia', 'James', 'Ava', 'Benjamin', 'Isabella'],
    'Age': [29, 31, 26, 33, 28, 34, 25, 27, 35, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [80000, 90000, 75000, 85000, 70000, 95000, 82000, 78000, 92000, 88000],
    'Email': ['john@gmail.com', 'emily@gmail.com', 'michael@gmail.com', 'sophia@gmail.com', 'william@gmail.com', 'olivia@gmail.com', 'james@gmail.com', 'ava@gmail.com', 'benjamin@gmail.com', 'isabella@gmail.com']
}
new_df = pd.DataFrame(new_data)

# Merge the new data with the existing DataFrame
df = pd.concat([df, new_df], ignore_index=True)
```

```python
def complex_selections(df):
  """
  This function performs complex selections on the DataFrame.
  """
  # People who live in New York
  new_york_residents = df[df['City'] == 'New York']

  # People who are younger than 35 and earn more than 80000
  young_rich = df[(df['Age'] < 26) & (df['Salary'] > 60000)]

  return new_york_residents, young_rich

# Call the function and display the results
new_york_residents, young_rich = complex_selections(df)

print("New York Residents:")
display(new_york_residents)

print("\nYoung and Rich:")
display(young_rich)
```

```python
def analyze_salary_by_city(df):
  """
  Analyzes the average salary by city.
  """
  return df.groupby('City')['Salary'].mean().sort_values(ascending=False)

def get_average_salary_and_age(df):
  """
  Calculates the average salary and age.
  """
  return df[['Salary', 'Age']].mean()

def find_employees(df, city=None, min_age=None, min_salary=None):
  """
  Finds employees who meet certain criteria.
  """
  filtered_df = df.copy()
  if city:
    filtered_df = filtered_df[filtered_df['City'] == city]
  if min_age:
    filtered_df = filtered_df[filtered_df['Age'] >= min_age]
  if min_salary:
    filtered_df = filtered_df[filtered_df['Salary'] >= min_salary]
  return filtered_df

def update_salaries(df, percentage, city=None):
  """
  Updates salaries by a certain percentage.
  """
  if city:
    df.loc[df['City'] == city, 'Salary'] *= (1 + percentage / 100)
  else:
    df['Salary'] *= (1 + percentage / 100)
  return df

# Example usage:
print("Average salary by city:")
display(analyze_salary_by_city(df))

print("\nAverage salary and age:")
display(get_average_salary_and_age(df))

print("\nEmployees in New York older than 25 with salary > 60000:")
display(find_employees(df, city='New York', min_age=25, min_salary=60000))

print("\nUpdate salaries in New York by 5%:")
df = update_salaries(df, 5, city='New York')
display(df)
```
